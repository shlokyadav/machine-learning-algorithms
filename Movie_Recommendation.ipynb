```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
```

```python
credits = pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
credits_r = credits.rename(index=str, columns={'movie_id':'id'})
merged = movies.merge(credits_r, on='id')
merged.head()
```

```python
merged.columns
```

```python
movies_c = merged.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_c.head()
```

```python
movies_c.columns
```

```python
movies_c.head(1)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode',analyzer='word'
                        ,token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')

movies_c['overview'] = movies_c['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(movies_c['overview'])
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(movies_c.index, index=movies_c['original_title']).drop_duplicates()
```

```python
indices
```

```python
indices["John Carter"]
```

```python
list(enumerate(sig[indices['John Carter']]))
```

```python
sorted(list(enumerate(sig[indices['John Carter']])), key=lambda x: x[1], reverse=True)
```

```python
def give_rec(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return movies_c['original_title'].iloc[movie_indices]
```

```python
give_rec('John Carter')
```
